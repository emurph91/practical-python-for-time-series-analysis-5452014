# Temporal resampling

## Load data

```python
path = '../../data/FRED/scripts/FRED_Data/FRED_joined.parquet'
```

```python
import pandas as pd
df = pd.read_parquet(path)

df
```

```python
df.plot()
```

## Steps

### Resample

```python
df.resample?
```

```python
r = df.resample(rule = "MS").agg(func = "mean")
```

### Visualize

#### Plot

```python
df.plot()
```

```python
r.plot()
```

#### Heatmatrix

```python
(r.style
.background_gradient(cmap="RdYlGn", axis=None)
.format(precision=2)
.format_index(formatter= lambda x: x.strftime('%Y'))

)
```

#### Heatmatrix with plotly

```python
r= r.T
```

```python
ranking = r.mean(axis=1).sort_values(ascending=False).index.tolist()
```

```python
r= r.loc[ranking]
```

```python
import plotly.express as px
px.imshow(img = r)
```
